```python
pontos = 0

while True:
    
    window, event, value = sg.read_all_windows()
    
    if window == janela0 and event == sg.WIN_CLOSED:
        break
    

    if window == janela0 and event == 'Bora':
        janela1 = pergunta1()
        janela0.hide()
    elif window == janela1:
        if value['c'] == True:
            pontos += 1
        else:
            pontos += 0

 
    if window == janela1 and event == 'Próximo':
        janela2 = pergunta2()
        janela1.hide()
    elif window == janela2:
        if value['b'] == True:
            pontos += 1
        else:
            pontos += 0
    

    if window == janela2 and event == 'Próximo':
        janela3 = pergunta3()
        janela2.hide()
    elif window == janela3:
        if value['d'] == True:
            pontos += 1
        else:
            pontos += 0
    

    if window == janela3 and event == 'Próximo':
        janela4 = pergunta4()
        janela3.hide()
    elif window == janela4:
        if value['a'] == True:
            pontos += 1
        else:
            pontos += 0
    
 
    if window == janela4 and event == 'Próximo':
        janela5 = pergunta5()
        janela4.hide()
    elif window == janela5:
        if value['b'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela5 and event == 'Próximo':
        janela6 = pergunta6()
        janela5.hide()
    elif window == janela6:
        if value['c'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela6 and event == 'Próximo':
        janela7 = pergunta7()
        janela6.hide()
    elif window == janela7:
        if value['d'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela7 and event == 'Próximo':
        janela8 = pergunta8()
        janela7.hide()
    elif window == janela8:
        if value['a'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela8 and event == 'Próximo':
        janela9 = pergunta9()
        janela8.hide()
    elif window == janela9:
        if value['d'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela9 and event == 'Próximo':
        janela10 = pergunta10()
        janela9.hide()
    elif window == janela10:
        if value['c'] == True:
            pontos += 1
        else:
            pontos += 0


    if window == janela10 and event == 'Voltar ao Início':
        janela0.un_hide()
        janela10.hide()    
    elif window == janela10 and event == 'Próximo':
        if pontos == 0:
            sg.popup(f'{pontos} Nenhum Acerto que horror')
        elif pontos >= 1 and pontos < 2:
            sg.popup(f'{pontos} Quase nenhum acerto, tem que melhorar')
        elif pontos >= 3 and pontos < 4:
            sg.popup(f'{pontos} Foi Bem, Bem Mau')
        elif pontos >= 5 and pontos < 6:
            sg.popup(f'{pontos} Na média, busque a perfeição')
        elif pontos >= 7 and pontos < 9:
            sg.popup(f'{pontos} O Conhecimento anda ao seu lado')
        elif pontos == 10:
            sg.popup(f'{pontos} Você é a Perfeição ')
        
    

janela0.close()
```
